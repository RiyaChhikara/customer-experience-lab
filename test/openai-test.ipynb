```python
import os
from dotenv import load_dotenv
from openai import OpenAI

load_dotenv()

client = OpenAI(api_key=os.getenv('OPENAI_API_KEY'))

# Sample complaint from Yelp
complaint = """
Called for emergency plumbing. They said they'd be there in 30 minutes. 
Waited 3 HOURS. When he finally showed up, he was rude and said it would 
cost $800 just to look at it. Then tried to upsell me on a full pipe 
replacement for $5000. My basement was flooding and he's trying to scam me!
"""

# Generate persona
response = client.chat.completions.create(
    model="gpt-4",
    messages=[{
        "role": "system",
        "content": "You are an expert at creating realistic customer personas from complaints. Generate a detailed persona with: name, age, situation, emotional state, specific frustration, and what they want resolved."
    }, {
        "role": "user",
        "content": f"Create a customer persona from this complaint:\n\n{complaint}"
    }],
    temperature=0.7
)

persona = response.choices[0].message.content
print("ðŸŽ­ Generated Persona:\n")
print(persona)
```
